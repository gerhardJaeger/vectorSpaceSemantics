# What is meaning?

 Semantics is the branch of linguistics dealing with meaning. But what is meaning?
 
> "In order to say what a meaning *is*, we may first ask what a meaning *does*, and then find something that does that." 
>
> (David Lewis, 1976, General semantics. In: Montague grammar. Academic Press, 1976. S. 1-50.)
 
 
 Broadly speaking, there are three philosophical conceptions of the nature of meanings:
 
 - truth-conditional semantics
 - cognitive semantics
 - usage-based semantics

## Truth-conditional semantics

Max Cresswell

> "Most Certain Principle: For two sentences $\alpha$ and $\beta$, if $\alpha$ is true and $\beta$ is false, $\alpha$ and $\beta$ must have different meanings."
>
> (Cresswell, M. J. 1982. The autonomy of semantics. In Stanley Peters & Esa Saarinen (eds.), Processes,
beliefs, and questions, 69–86. Reidel.)

Truth-conditional semantics strengthens this to a *Less certain principle*: 

*If $\alpha$ and $\beta$ have the same truth conditions, they have the same meaning.*

This is very close to the prevalent notion of meaning in formal logic. *Formal semantics* starts from the less certain principle plus the *Principle of compositionality*:

*The meaning of a complex expression is determined by the meaning of its parts and the way they are combined.*

The focus of research in formal semantics is on the composition part, while the meanings of individual words are treated as black box.

## Cognitive semantics

Cognitive semantics, simply speaking, assumes that *meanings are in the head*. This turns semantics into a branch of psychology and perhaps neuro-science.

A major challenge for cognitive semantics is to give **identity criteria** for meanings. It stands to reason that a prolific author has a different conception of *book* than somebody who rarely reads. Still, we would probably want to insist that there is a inter-subjective meaning of that word &ndash; the kind of information you would find in a dictionary &ndash;, which is accessible to both people.

Most cognitivist approaches identify meanings with positions in an association network.

## Usage-based semantics

- **"Die Bedeutung eines Wortes liegt in seinem Gebrauch"** (Ludwig Wittgenstein, Philosophical Investigations)


- **"You shall know a word by the company it keeps!"** (J. R. Firth, A synopsis of linguistic theory 1930–55. In Studies in linguistic analysis, pages 1–32. The Philological Society, Oxford)


- **"What people know when they say that they know a word is not how to recite its dictionary definition &ndash; they know how to use it [...] in everyday discourse."** (Miller, George A. (1986). Dictionaries in the mind. Language and Cognitive Processes, 1, 171–185.)


- **Distributional hypothesis: difference of meaning correlates with difference of distribution** (Zellig Harris)

## Usage-based semantics

- *vector space semantics* is a version of usage-based semantics

- strengths of this approach:
   - gradient notion of semantic similarity
   - principled approach how meanings can be learned

- challenges:
   - predict what meanings do (remember the Lewis quotation above)


# Desiderata for a theory of meaning

## meaning relations

### synonymy

- *buy* – *purchase*

- *sick* – *ill*

- *cell phone* – *mobile phone*


### hyponomy

- *fir* – *conifer*
- *beetle* – *insect*
- *water* – *liquid*



### antonymy

- *small* – *large*
- *thick* – *thin*
- *rich* – *poor*

### semantic complementarity

- *even* – *odd*
- *unmarried* – *married*
- *on* – *off*



### homonymy

- *bank* (financial institution – edge of a river)
- *band* (ribbon – group of musicians)
- *bow* (weapon to shoot projectiles – kind of tied ribbon)

### polysemy

- *bank* (financial institution – building where the institution has a branch – entirety of employees of the financial institution)

## connotations

cultural and/or emotional associations carried by an expression, over and above its descriptive meaning

- *strong-willed* – *pig-headed*

- *childish* – *youthful*

- *skinny* – *lithe* – *thin*

```python
import numpy as np
import pandas as pd

M = pd.DataFrame(
    np.array([
        [51, 20, 84, 0, 3, 0],
        [52, 58, 4, 4, 6, 26],
        [115, 83, 10, 42, 33, 17],
        [59, 39, 23, 4, 0, 0],
        [98, 14, 6, 2, 1, 0],
        [12, 17, 3, 2, 9, 27],
        [11, 2, 2, 0, 18, 0]
    ]),
    ['knife', 'cat', 'dog', 'boat', 'cup', 'pig', 'banana'],
    ['get', 'see', 'use', 'hear', 'eat', 'kill']
)
```

**co-occurrence matrix** $M$

```python
M
```

- row-vector $x_{\textrm{dog}}$ describes usage of the word *dog* in the corpus

```python
M[2:3]
```

- can be seen as coordinates of point in $n$-dimensional Euclidean space

Illustrated for dimensions *get* and *use*:

```python
import matplotlib.pyplot as plt
from matplotlib import cm
import string
cmap = cm.get_cmap('Spectral')
```

```python
M1 = M[['get','use']][:4]
M1
```

```python
fig, ax = plt.subplots()
M1[['get', 'use']].plot('get', 'use', 
                        kind='scatter', 
                        ax=ax, s=120, linewidth=0, c=range(len(M1)), colormap=cmap)
for k, v in M1[['get', 'use']].iterrows():
    ax.annotate(k, v,
                xytext=(10,-5),
                textcoords='offset points', 
                family='sans-serif', 
                fontsize=18, 
                color='black')
plt.xlim([0,130])
plt.ylim([0,120])
```

Spatial proximity is expected to correspond to semantic similarity.

```python
from scipy.spatial import distance

distance.euclidean(M1.loc['dog'].values, M1.loc['cat'].values)
```

```python
distance.euclidean(M1.loc['dog'].values, M1.loc['boat'].values)
```

```python
distance.euclidean(M1.loc['dog'].values, M1.loc['knife'].values)
```

This doesn't work too well here because the position of a word depends on its frequency of occurrence, which affects the Euclidean distance.

Other distance measure give better results.

```python
distance.cosine(M1.loc['dog'].values, M1.loc['cat'].values)
```

```python
distance.cosine(M1.loc['dog'].values, M1.loc['boat'].values)
```

```python
distance.cosine(M1.loc['dog'].values, M1.loc['knife'].values)
```

```python
wordlists = np.array([np.array(nltk.tokenize.word_tokenize(s)) for s in df.Text])
```

```python
freq_dist = nltk.FreqDist()
```

```python
for sentence in wordlists:
    for w in sentence:
        freq_dist[w.lower()] += 1
```
